```python
import pandas as pd

df = pd.read_csv('Numeros1.csv', encoding='ISO-8859-1')
df
```

```python
# GrÃ¡fico de dispersÃ£o

df.plot(kind="scatter", x="consumo", y="valor" );
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Criando um modelo de Machine Learning

modelo = linear_model.LinearRegression()
```

```python
x = df[['consumo']]
y = df[['valor']]
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)
```

```python
# Treinando o modelo de Machine Learning

modelo.fit(x_treino, y_treino)
```

```python
previsoes = modelo.predict(x_teste)
```

```python
# Testando o modelo e a porcentagem de acerto nas previsÃµes

acuracidade = round(r2_score(y_teste, previsoes) * 100, 2)
print(f'A acuracidade do modelo Ã© de {acuracidade}%')
```

```python
import pickle

with open("modelo_treinado.pkl", "wb") as file:
    pickle.dump(modelo, file)
```
